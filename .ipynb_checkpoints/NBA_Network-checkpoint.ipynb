
# Social Network Analysis of the NBA

```python
# import libraries
import csv
import math
import networkx as net
from networkx.algorithms import bipartite as bi
import matplotlib as plt
import pandas as pd
from matplotlib.pyplot import figure
import numpy as np
```

```python
# open nba data as a CSV file
nba_data = csv.reader(open("NBA_network_all_v2.csv"))
nba_data_df = pd.read_csv("NBA_network_all_v2.csv", header = None,encoding ='utf-8')
```

```python
# create network 
g = net.Graph()

# lists of all distinct teams and players
teams = []
players = []

# construct directed graph
for row in nba_data:
    if row[0] not in teams:
        teams.append(row[0])
    if row[1] not in players:
        players.append(row[1])
    g.add_edge(row[0], row[1], weight=1)
```

```python
# here we generate a distinct player list       
with open('players.txt', 'w') as f:
    for player in players:
        f.write(player + '\n')
```

```python
# here were we generate skill matrix needed to implement the two algorithms from paper1 
unique_players = pd.DataFrame(columns=['Player'])
for player in players:
    if player not in unique_players['Player'].values:
        unique_players = unique_players.append({'Player': player}, ignore_index=True)
positions = {}
for index, row in nba_data_df.iterrows():
    player = row[1]
    position = row[6]
    if player in positions:
        if positions[player] != position:
            positions[player] += '/' + position
    else:
        positions[player] = position
matrix = np.zeros((len(unique_players), 5))

# Iterate over the unique player names and update the matrix based on the positions played
for index, row in unique_players.iterrows():
    player = row['Player']
    if player in positions:
        position_list = positions[player].split('/')
        for position in position_list:
            if position == 'SG':
                matrix[index][0] = 1
            elif position == 'PG':
                matrix[index][1] = 1
            elif position == 'SF':
                matrix[index][2] = 1
            elif position == 'PF':
                matrix[index][3] = 1
            elif position == 'C':
                matrix[index][4] = 1

# Convert the numpy array to a pandas dataframe
final_df = pd.DataFrame(matrix, columns=['SG', 'PG', 'SF', 'PF', 'C'])

# Merge the unique_players dataframe and final_df dataframe to get the final result
result1 = pd.concat([unique_players, final_df], axis=1)
# result1 is a matrix that assigns its cell value=1 , if the player has played that specific position.
```

```python
#here we store the result1 matrix obtained from above in csv file fromat
result1.to_csv("result1.csv", index=False)
```

```python
#here we store the result1 matrix obtained from above in .mat file fromat
import scipy.io as sio
sio.savemat('count_label_nba.mat', {'result1': result1})
```

# Below code is for generating the NBA social network

```python
print ("nodes in network:", len(g))
```

```python
connected_net = list(net.connected_components(g))
print ("size of largest component:", len(connected_net[0]))
```

```python
# players with the most seasons
print ("Players with highest counts (not most seasons):")
print (nba_data_df[1].value_counts().head(10))
```

```python
# create an affiliation network of only players
# this will become our main network..
playernet = bi.weighted_projected_graph(g, players, ratio = False)
```

```python
# here we store the network in an matrix form using numpy function. 
adj_matrix_nba = net.to_numpy_array(playernet)
```

Here , what we want is generate a player network , such that each node in the network is the players name and the weight of edge between the players is total number of seasons they played together as teamates.

```python
import pandas as pd

# Convert the numpy array to a pandas dataframe
df = pd.DataFrame(adj_matrix_nba)

# Write the dataframe to an Excel file
df.to_excel("adj_matrix_nba.xlsx", index=False)
```

```python
#here we store the adj_matrix_nba matrix obtained from above in .mat file fromat
import scipy.io as sio
sio.savemat('adj_matrix_nba.mat', {'adj_matrix_nba': adj_matrix_nba})
```

```python
# list of the players with most seasons as teammates:
connect_list = list(playernet.edges(data=True))
player_connections = sorted(connect_list, key=lambda tup: tup[2]['weight'])
player_connections = player_connections[::-1]
print ("Players with most seasons as teammates:")
player_connections[0:20]
```
